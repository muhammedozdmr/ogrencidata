```
import numpy as np
import pandas as pd
#Excel Dosyası Okuma
dataset = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/pisa2009test.xlsx')
#Veri Tiplerinin Çıktısı
print(dataset.dtypes)
dataset
```

**DATASET DEĞİŞKENİNDEKİ VERİLERİN X VE y DEĞİŞKENLERİNE ATANIP YAZDIRILMASI**

```
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(X)
print(y)
```

**NaN VERİ DÖNÜŞÜMÜ / DÖNÜŞÜM SONRASI X DEĞİŞKENİNİ YAZDIRMA**

```
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(X[:,2:25])
X[:,2:25]=imp.transform(X[:,2:25])
```

**DÖNÜŞÜMÜ SAĞLANAN NaN VERİLERLE BİRLİKTE YENİ X DİZİSİNİN ÇIKTISI**

```
print(X)
```

```
print(X[1567])
```

**OBJECT DEĞERLERİN ONEHOTENCODING İLE DÖNÜŞÜMÜ**

```
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encode',OneHotEncoder(),[1])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
print(X[1567])
```

**y ETİKET SÜTUNUNU BINARY POZİSYONUNA GETİRMEK İÇİN LABEL DÖNÜŞÜMÜ YAPMA**

> **NOT: ETİKET SÜTUNUMUZ ZATEN BINARY POZİSYONUNDA**




```
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

**ELDE EDİLEN X VE y DİZİLERİNE EĞİTİM VE TEST İŞLEMİ UYGULANIP EĞİTİM VERİLERİNİN ELDE EDİLMESİ**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

print(X_train)
print(X_train[0])
```
